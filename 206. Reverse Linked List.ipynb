# Very Important - Reverse Linked List

Given the `head` of a singly linked list, reverse the list, and return the *reversed* list.



# Solution 


There are 2 types of approach to this type of problem. First we will use the recursive approach, and then the iterative approach

```python
from typing import Optional
```

```python
# Iterative Approach

class Solution:
    def reverseList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        newHead = None               # "newHead" is to be returned. By now it's tail.
        while head:
            previous = newHead
            newHead = head           # Now any doing with "newHead" makes the same doing with "head".
            head = head.next         # !!! That's why this line can't be placed after the next.  
            newHead.next = previous  # Since "head" is freed, "newHead.next = previous" will not cause "head.next = previous".
        return newHead
```
